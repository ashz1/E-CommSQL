```python
import pandas as pd

# Load the datasets
flipkart_df = pd.read_excel('data/1.eTailing_Flipkart.xlsx')
amazon_df = pd.read_excel('data/2. eTailing_Amazon.xlsx')


# Remove leading/trailing spaces from column names
flipkart_df.columns = flipkart_df.columns.str.strip()
amazon_df.columns = amazon_df.columns.str.strip()

# Check for any missing values
print("\nMissing Values in Flipkart Dataset:")
print(flipkart_df.isnull().sum())

print("\nMissing Values in Amazon Dataset:")
print(amazon_df.isnull().sum())

# Display the cleaned column names and first few rows
print("Cleaned Flipkart Columns:")
print(flipkart_df.columns)
print(flipkart_df.head())

print("\nCleaned Amazon Columns:")
print(amazon_df.columns)
print(amazon_df.head())

# Ensure columns align correctly for merging
common_columns = list(set(flipkart_df.columns).intersection(set(amazon_df.columns)))

# Filter the columns of both dataframes to include only the common columns
flipkart_df = flipkart_df[common_columns]
amazon_df = amazon_df[common_columns]



```

```python
import sqlite3

# Create a connection to a new SQLite database
conn = sqlite3.connect('etailing_data.db')

# Create tables
flipkart_df.to_sql('flipkart_data', conn, if_exists='replace', index=False)
amazon_df.to_sql('amazon_data', conn, if_exists='replace', index=False)

# Close the connection
conn.close()
```

```python
import sqlite3

conn = sqlite3.connect('etailing_data.db')
cursor = conn.cursor()

# Example query: Compare total revenue between Flipkart and Amazon
query = """
SELECT 
    'Flipkart' as Company,
    SUM("Total Revenue (USD Mn)") as Total_Revenue
FROM flipkart_data
UNION ALL
SELECT 
    'Amazon' as Company,
    SUM("Total Revenue (USD Mn)") as Total_Revenue
FROM amazon_data
"""

cursor.execute(query)
results = cursor.fetchall()

for row in results:
    print(f"{row[0]}: {row[1]}")

conn.close()
```
